```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = df.copy()

# Define variables of interest
clinical_variables = {
    "headache_reported": "Headache Reported",
    "amnesia_present": "Amnesia Present",
    "altered_mental_status": "Altered Mental Status",
    "posttraumatic_seizure": "Posttraumatic Seizure",
    "vomiting_reported": "Vomitting",
    "palpable_skull_fracture": "Palpable Skull Fracture",
    "basilar_skull_fracture_signs": "Basilar Skull Fracture",
    "scalp_hematoma_present": "Scalp Hematoma Present",
    "trauma_above_clavicles": "Trauma Above Clavicles",
    "neurological_deficit_present": "Neurological Deficit",
    "loss_of_consciousness": "Loss of Consciousness",
}

ct_reasons = {
    "head_ct_planned": "Head CT Planned",
    "ct_reason_age": "CT Reason: Age",
    "ct_reason_amnesia": "CT Reason: Amnesia",
    "ct_reason_ams": "CT Reason: Altered Mental Status",
    "ct_reason_clinical_skull_fracture": "CT Reason: Clinical Skull Fracture",
    "ct_reason_headache": "CT Reason: Headache",
    "ct_reason_scalp_hematoma": "CT Reason: Scalp Hematoma",
    "ct_reason_loss_of_consciousness": "CT Reason: Loss of Consciousness",
    "ct_reason_mechanism": "CT Reason: Mechanism of Injury",
    "ct_reason_neurological_deficit": "CT Reason: Neurological Deficit",
    "ct_reason_request_md": "CT Reason: Physician Request",
    "ct_reason_request_parent": "CT Reason: Parent Request",
    "ct_reason_request_trauma": "CT Reason: Trauma Team Request",
    "ct_reason_seizure": "CT Reason: Seizure",
    "ct_reason_vomiting": "CT Reason: Vomiting",
    "ct_reason_xray_skull_fracture": "CT Reason: Skull Fracture on X-Ray",
    "ct_reason_other": "CT Reason: Other",
}

# Prepare age categories
age_labels = {1: "Under 2", 2: "2 and above"}
age_categories = sorted(data["age_category"].dropna().unique())
totals = {cat: len(data[data["age_category"] == cat]) for cat in age_categories}


def calculate_percentages(data, variables):
    results = []
    for col, display_name in variables.items():
        for age_cat in age_categories:
            subset = data[data["age_category"] == age_cat]
            count = (subset[col] == 1).sum()
            percentage = (count / totals[age_cat] * 100) if totals[age_cat] > 0 else 0
            results.append(
                {
                    "Variable": display_name,
                    "age_category": age_cat,
                    "percentage": percentage,
                }
            )
    return pd.DataFrame(results)


def prepare_plot_data(df_results):
    pivot_data = df_results.pivot(
        index="Variable", columns="age_category", values="percentage"
    ).fillna(0)
    pivot_data["Total"] = pivot_data.sum(axis=1)
    pivot_data = pivot_data.sort_values("Total", ascending=False)

    plot_data = (
        pivot_data.drop("Total", axis=1)
        .reset_index()
        .melt(id_vars="Variable", var_name="age_category", value_name="percentage")
    )
    plot_data["age_category"] = plot_data["age_category"].map(age_labels)
    plot_data["Variable"] = pd.Categorical(
        plot_data["Variable"], categories=pivot_data.index, ordered=True
    )
    return plot_data


def create_subplot(ax, data, title, ylabel):
    sns.barplot(
        data=data,
        x="percentage",
        y="Variable",
        hue="age_category",
        hue_order=[age_labels[cat] for cat in age_categories],
        palette="viridis",
        ax=ax,
    )
    ax.set_xlabel("Percentage of Patients")
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.legend(title="Age Category", loc="lower right")

    for patch in ax.patches:
        width = patch.get_width()
        if width > 0:
            ax.text(
                width + 0.5,
                patch.get_y() + patch.get_height() / 2,
                f"{width:.1f}%",
                va="center",
                fontsize=8,
            )


# Calculate and prepare data for both plots
clinical_data = prepare_plot_data(calculate_percentages(data, clinical_variables))
ct_data = prepare_plot_data(calculate_percentages(data, ct_reasons))

# Create visualization
fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(20, 10))
create_subplot(
    axes[0], clinical_data, "Clinical Variables by Age Category", "Clinical Variables"
)
create_subplot(axes[1], ct_data, "CT Reasons by Age Category", "CT Reasons")

plt.tight_layout()
plt.savefig("../figs/age_category.svg", format="svg", bbox_inches="tight")
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = df.copy()

# Define variables of interest
columns_of_interest = {
    "loss_of_consciousness": "Loss of Consciousness",
    "headache_reported": "Headache Reported",
    "vomiting_reported": "Vomiting Reported",
    "intubation_status": "Intubation Status",
    "altered_mental_status": "Altered Mental Status",
    "basilar_skull_fracture_signs": "Basilar Skull Fracture",
    "gcs_total": "GCS Total",
    "head_ct_planned": "Head CT Planned",
    "clinically_important_tbi": "Clinically Important TBI",
}

# Prepare data for correlation analysis
correlation_data = data[columns_of_interest.keys()].replace(
    {90: np.nan, 91: np.nan, 92: np.nan}
)
correlation_data = correlation_data.rename(columns=columns_of_interest)

# Calculate original correlation matrix
original_correlation = correlation_data.corr(method="spearman")

# Create perturbed dataset
perturbed_data = correlation_data.copy()
num_columns = perturbed_data.shape[1]
columns_to_perturb = max(1, int(0.10 * num_columns))

for row_idx in range(len(perturbed_data)):
    selected_columns = np.random.choice(
        range(num_columns), size=columns_to_perturb, replace=False
    )
    for col_idx in selected_columns:
        column_name = perturbed_data.columns[col_idx]
        valid_values = correlation_data[column_name].dropna().unique()
        if len(valid_values) > 0:
            perturbed_data.iloc[row_idx, col_idx] = np.random.choice(valid_values)

# Calculate perturbed correlation matrix
perturbed_correlation = perturbed_data.corr(method="spearman")

# Create visualization
fig, axes = plt.subplots(ncols=2, figsize=(20, 8))

# Plot original correlation matrix
sns.heatmap(
    original_correlation,
    cmap="viridis",
    annot=True,
    fmt=".2f",
    linewidths=0.5,
    ax=axes[0],
)
axes[0].set_title("Original Correlation Matrix (Spearman)")

# Plot perturbed correlation matrix
sns.heatmap(
    perturbed_correlation,
    cmap="viridis",
    annot=True,
    fmt=".2f",
    linewidths=0.5,
    ax=axes[1],
)
axes[1].set_title("Perturbed Correlation Matrix (10% Values Perturbed)")

plt.tight_layout()
plt.savefig(
    "../figs/correlation_matrix_perturbed.svg", format="svg", bbox_inches="tight"
)
plt.show()
```
