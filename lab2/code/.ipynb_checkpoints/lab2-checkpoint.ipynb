```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
import glob
```

# Part 1: EDA
### 1.1

```python
# Loading the 3 expert label files
files = ['O013257.npz', 'O013490.npz', 'O012791.npz']
data = []

for i in files:
    file_path = os.path.join("../data/image_data", i)
    npz_data = np.load(file_path)
    key = list(npz_data.files)[0]
    data.append(npz_data[key])
    print("loaded", i, "with shape", npz_data[key].shape)
```

```python
# Creating the plot figure
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Creating values for the colors so 1 (cloud) is blue, 
# -1 is red (non-cloud) and 0 is grey (unlabeled)
colors = {1: 'blue', -1: 'red', 0: 'grey'} 

# Open and loop through each of the 3 images
for idx, (ax, file_data) in enumerate(zip(axes, data)):
    # Get the x coordinatee from the first column
    x = file_data[:, 1]
    # Get the y coordinate from column 0
    y = file_data[:, 0]
    # Get the expert labels from the 11th column
    labels = file_data[:, 10]
    
    # Scatter plot for each image
    for i, j in colors.items():
        mask = (labels == i)
        ax.scatter(x[mask], y[mask], c=j, s=1, alpha=0.7, label='Cloud' if i == 1 else 'Non-cloud' if i == -1 else 'Unlabeled')
    
    # Setting up the legend, title and labels
    ax.set_title(f'Image {files[idx]}')
    ax.set_xlabel('X Coordinate')
    ax.set_ylabel('Y Coordinate')
    ax.legend()

plt.tight_layout()
plt.show()

# Each of the graphs show the cloud non cloud and unlabeled images
```

### 1.2

```python
data_columns = ['y', 'x', 'NDAI', 'SD', 'CORR', 'DF', 'CF', 'BF', 'AF', 'AN', 'expert_label']
labeled_data = []

for idx, file_data in enumerate(data):
    # Filter out just the cloud and non-cloud pixels and ignore unlabeled pixels
    mask = (file_data[:, 10] == 1) | (file_data[:, 10] == -1)
    filtered = file_data[mask]
    
    # Convert the filtered array to pd dataframe
    df = pd.DataFrame(filtered, columns=data_columns)
    
    # Adding to know which image the data came from
    df['image'] = files[idx]
    
    # Appending the dataframe to the list of labeled data
    labeled_data.append(df)
    
    print(files[idx], "has", df.shape[0], "labeled pixels")

total_labeled = pd.concat(labeled_data, ignore_index=True)
print("total labeled", total_labeled.shape[0])
```

```python
# Plot cloud vs non-cloud radiances relationshop
plot = sns.pairplot(
    total_labeled,
    vars=['DF', 'CF', 'BF', 'AF', 'AN'],
    hue='expert_label',
    palette={1: 'royalblue', -1: 'crimson'},
    plot_kws={'alpha': 0.6, 's': 12, 'edgecolor': 'none'},
    diag_kind='kde'
)

plt.suptitle("Pairwise Relationships: Cloud vs. Non-Cloud Radiances", y=1.03, fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Compute correlation matrices for clouds and non-clouds separately
cloud_matrix = total_labeled[total_labeled['expert_label'] == 1][['DF', 'CF', 'BF', 'AF', 'AN']].corr()
non_cloud_matrix = total_labeled[total_labeled['expert_label'] == -1][['DF', 'CF', 'BF', 'AF', 'AN']].corr()

# Plot correlation matrices for cloud pixels
# Used seaborn heatmap to visualize positive/negative correlations
plt.figure(figsize=(8, 6))
sns.heatmap(cloud_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, annot_kws={"size": 8})
plt.title("Correlation of Radiance Angles (Cloud Pixels)")
plt.tight_layout()
plt.show()

# Plot correlation matrices for non-cloud pixels 
# Used seaborn heatmap to visualize positive/negative correlations
plt.figure(figsize=(8, 6))
sns.heatmap(non_cloud_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, annot_kws={"size": 8})
plt.title("Correlation of Radiance Angles (Non-Cloud Pixels)")
plt.tight_layout()
plt.show()
```

```python
# Subset the data to only include NDAI, SD, CORR and expert label
features = total_labeled[['expert_label', 'NDAI', 'SD', 'CORR']]

# Mean, std, min, max grouped by expert label for NDAI, SD, CORR
basic_stats = features.groupby('expert_label').agg(['mean', 'std', 'min', 'max'])
print(basic_stats)
# Used a pairplot to visualize the relationships between NDAI, SD, CORR and expert label
sns.pairplot(features, vars=['NDAI','SD','CORR'], hue='expert_label',
             palette={1: 'blue', -1: 'red'}, plot_kws={'alpha': 0.4, 's': 10})
plt.suptitle("Pairwise plot of NDAI, SD, CORR by expert label", y=1.02)
plt.show()
```

```python
# Separate data by expert label
cloud_data = total_labeled[total_labeled['expert_label'] == 1]
noncloud_data = total_labeled[total_labeled['expert_label'] == -1]

# reference: https://stackoverflow.com/questions/23199796/detect-and-exclude-outliers-in-a-pandas-dataframe

# Handle outliers separately for each class
def outliers_function(df, features_list, z_threshold=3):
    # Calculate z-scores for features
    z_scores = np.abs((df[features_list] - df[features_list].mean()) / df[features_list].std())
    
    # Create masks for different features with different thresholds
    ndai_outliers = z_scores['NDAI'] > z_threshold
    sd_outliers = z_scores['SD'] > z_threshold
    corr_outliers = z_scores['CORR'] > z_threshold
    
    outliers = (ndai_outliers + sd_outliers + corr_outliers) > 0

    # Cleaned data and outliers
    return df[~outliers], df[outliers]

# For each class
cleaned_cloud, outlier_cloud = outliers_function(cloud_data, ['NDAI', 'SD', 'CORR'])
cleaned_noncloud, outlier_noncloud = outliers_function(noncloud_data, ['NDAI', 'SD', 'CORR'])

# Combine cleaned data
cleaned_data = pd.concat([cleaned_cloud, cleaned_noncloud])

# Calculate statistics
total_outliers = len(outlier_cloud) + len(outlier_noncloud)
outlier_percentage = (total_outliers / len(total_labeled)) * 100

# Summary statistics
print(f"Total outliers detected: {total_outliers} out of {len(total_labeled)} pixels")
print(f"Outlier percentage: {outlier_percentage:.2f}%")
print(f"Cloud pixels outliers: {len(outlier_cloud)} ({len(outlier_cloud)/len(cloud_data)*100:.2f}%)")
print(f"Non-cloud pixels outliers: {len(outlier_noncloud)} ({len(outlier_noncloud)/len(noncloud_data)*100:.2f}%)")

# Print statistics about each feature for both classes
for feature in ['NDAI', 'SD', 'CORR']:
    # Original data stats
    print(f"\n - {feature}")
    print(f"Original data range: {total_labeled[feature].min():.3f} to {total_labeled[feature].max():.3f}")
    
    # Cleaned data statistics
    print(f"Cleaned data range: {cleaned_data[feature].min():.3f} to {cleaned_data[feature].max():.3f}")
    
    # Statistics by class
    print(f"\nCloud pixels - Before cleaning: {cloud_data[feature].mean():.3f}±{cloud_data[feature].std():.3f}")
    print(f"Cloud pixels - After cleaning: {cleaned_cloud[feature].mean():.3f}±{cleaned_cloud[feature].std():.3f}")
    
    print(f"Non-cloud pixels - Before cleaning: {noncloud_data[feature].mean():.3f}±{noncloud_data[feature].std():.3f}")
    print(f"Non-cloud pixels - After cleaning: {cleaned_noncloud[feature].mean():.3f}±{cleaned_noncloud[feature].std():.3f}")
```

```python
# Check for missing values in the entire dataset
missing_values = total_labeled.isnull().sum()
print("\nMissing values in each column:")
print(missing_values)

# Check for  negative radiance values
neg_radiance_mask = (total_labeled['DF'] < 0) | (total_labeled['CF'] < 0) | \
                    (total_labeled['BF'] < 0) | (total_labeled['AF'] < 0) | \
                    (total_labeled['AN'] < 0)
invalid_pixels = total_labeled[neg_radiance_mask]
print(f"\nPixels with negative radiance values: {len(invalid_pixels)}")

#Look for zero radiance values 
zero_radiance_mask = (total_labeled['DF'] == 0) | (total_labeled['CF'] == 0) | \
                     (total_labeled['BF'] == 0) | (total_labeled['AF'] == 0) | \
                     (total_labeled['AN'] == 0)
zero_pixels = total_labeled[zero_radiance_mask]
print(f"Pixels with zero radiance values: {len(zero_pixels)}")

#Handle missing or invalid values 
valid_data_mask = ~neg_radiance_mask & ~zero_radiance_mask & ~total_labeled.isnull().any(axis=1)
valid_data = total_labeled[valid_data_mask]

print(f"\nTotal pixels after removing missing/invalid values: {len(valid_data)} out of {len(total_labeled)}")
print(f"Percentage of data retained: {len(valid_data)/len(total_labeled)*100:.2f}%")

# Combine both cleaning steps
final_cleaned_data = cleaned_data[~cleaned_data.index.isin(invalid_pixels.index) & 
                                 ~cleaned_data.index.isin(zero_pixels.index)]

print(f"\nFinal dataset size after all cleaning: {len(final_cleaned_data)}")
print(f"Total reduction: {(1 - len(final_cleaned_data)/len(total_labeled))*100:.2f}%")
```
